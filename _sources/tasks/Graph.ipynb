```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.DiGraph()
```

```python
edge_list = [
    (1, 3), (1, 4), (1, 5),
    (2, 3), (2, 5), (2, 6), 
    (3, 1), (3, 2), (3, 4), (3, 5),
    (4, 1), (4, 3), (4, 5), (4, 6),
    (5, 1), (5, 2), (5, 3), (5, 4), (5, 6),
    (6, 2), (6, 4), (6, 5),
    (7, 2), (7, 8),
    (8, 7),
    (9, 5), (9, 8),
]
```

```python
G.add_edges_from(edge_list)
```

```python
G = nx.relabel_nodes(G, {1: "A", 2: "B", 3: "C", 4: "D", 5: "E", 6: "F", 7: "G", 8: "H", 9: "I"})
```

```python
nx.draw_circular(G, with_labels=True)
plt.show()
```

```python
pr = nx.pagerank(G, alpha=0.85, max_iter=22)
```
